```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs 
import scipy.spatial.distance   # 距离函数 默认欧式距离
```

```python
X, y = make_blobs(n_samples=100, centers=6, random_state=1234, cluster_std=0.6)

plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# 实现K_Means
class K_Means(object):
    # 初始化 n_clusters(k)，centroids (初始质心) ， num_iter
    def __init__(self, n_clusters=6, centroids=[], num_iter=300):
        self.n_clusters=n_clusters
        self.num_iter=num_iter
        self.centroids=np.array( centroids,dtype=np.float)
    
    # 训练模型方法 k-means聚类过程
    def fit(self, X):
        # 假如没有指定质心，就随机从X中选k个
        if( self.centroids.shape!=(self.n_clusters,X.shape[1]) ):
            selt.centroids = X[ np.random.randint(0, high=X.shape[0], size = self.n_clusters), :] 
        # 开始迭代
        for i in range(self.num_iter):
            # 假如X有100行 ,6个质心  返回 100*6 代表   1*6 代表X中某行（点）与6个质心的距离
            dists = scipy.spatial.distance.cdist(X, self.centroids)
            # 100*1 代表距离每个样本最近的是 第几个质心
            c_ind = np.argmin( dists,axis=1)
            # 对每一类样本进行均值计算，更新质心
            for idx in range( self.n_clusters ):
                if( idx in c_ind ):
                    # 用属于同一个质心的点的均值，更新第idx个质心    
                    self.centroids[idx] = np.mean( X[c_ind==idx], axis=0 )  
                    
    def predict(self, samples):
        dists = scipy.spatial.distance.cdist(samples, self.centroids)
        c_ind = np.argmin( dists,axis=1)
        return c_ind
```

```python
# 测试
# 定义一个绘制子图函数
def plotKMeans(X, y, centroids, subplots, title):
    #分配子图，121表示1行2列子图的第一个
    plt.subplot(subplots)
    plt.scatter(X[:,0], X[:,1],c='r')
    # 画出质心
    plt.scatter(centroids[:,0], centroids[:,1], s=100, c=np.array( range(centroids.shape[0])))
```

```python
init_centroids = np.array([[2,1],[2,2],[2,3],[2,4],[2,5],[2,6]]) 
plt.figure(figsize=(16,6))
kmeans = K_Means(centroids = init_centroids)
plotKMeans(X, y, kmeans.centroids, 121, 'Inital State')

# 开始聚类
kmeans.fit(X)
plotKMeans(X, y, kmeans.centroids, 122, 'Final State')


# 预测新数据点的类别
x_new = np.array([[0,0],[10,7]])
y_pred = kmeans.predict(x_new)
print(y_pred)
plt.scatter(x_new[:,0],x_new[:,1],c='black',s=100)

kmeans.centroids[y_pred]
plt.
```
